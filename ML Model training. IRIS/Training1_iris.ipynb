```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
type(data)
```

```python
df=pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df
```

```python
df['target']=np.zeros(150)
df.head(10)
```

```python
df['target']=np.ones(150)
df.describe()
```

```python
"""
calculate standard daviation (std) for sepal length
"""
dif=(df['sepal length (cm)'] - df['sepal length (cm)'].mean())**2
np.sqrt(dif.sum()/149)
```

```python
dif=(df['petal length (cm)'] - df['petal length (cm)'].mean())**2
np.sqrt(dif.sum()/149)
```

```python
"""
plot(x-horizontal, y-vertical, connection type)
"""
plt.plot(df['petal length (cm)'],df['sepal length (cm)'],'.r')
```
