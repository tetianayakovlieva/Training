```python
import pandas as pd
import numpy as np
```

```python
data={'height':[170,184,179,189,150,147,165,187,171,166], 'weight':[70,100,82,91,45,41,51,68,71,80]}
```

```python
df=pd.DataFrame(data)
```

```python
df['body_index']=((df['height']-df['weight'])/100-1.05)*10
```

```python
df['true_gender']=(df['body_index']>=0).astype(int)
```

```python
df
```

## f(height,weight)->probability that we have woman or 1
## f(a,b,c,height,weight)
## f=logistic_function(a x height + b x weight + c)

```python
a=-5
b=-4
c=1000
df['logits']=a*df['height']+b*df['weight']+c
```

```python
df
```

```python
df['predicted_probability']=1/(1+1/np.exp(df['logits']))
```
