```python
import pandas as pd
import numpy as np
```

```python
url = ('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/tips.csv')
tips = pd.read_csv(url)
tips
```

```python
df = pd.DataFrame(tips)
df
```

```python
sorted_df = df.sort_values("total_bill")
sorted_df
```

```python
df.sort_values("total_bill", inplace=True)
df
```

SELECT total_bill, tip, smoker, time FROM tips;

```python
tips[["total_bill", "tip", "smoker", "time"]]
```

SELECT *, tip/total_bill as tip_rate FROM tips;

```python
tips.assign(tip_rate=tips["tip"]/tips["total_bill"])
```

SELECT * FROM tips WHERE time = 'Dinner';

```python
 tips[tips["total_bill"] > 10]
```

```python
is_dinner = tips["time"] == "Dinner"
is_dinner
```

```python
is_dinner.value_counts()
```

```python
tips[is_dinner]
```

SELECT * FROM tips WHERE time = 'Dinner' AND tip > 5.00;

```python
tips[(tips["time"] == "Dinner") & (tips["tip"] > 5.00)]
```

SELECT * FROM tips WHERE size >= 5 OR total_bill > 45;

```python
tips[(tips["size"] >= 5) | (tips["total_bill"] > 45)]
```

SELECT * FROM tips WHERE tip IS NULL;

```python
tips[tips["tip"].isna()]
```

SELECT * FROM tips WHERE tip IS NOT NULL;

```python
tips[tips["tip"].notna()]
```

SELECT sex, count(*) FROM tips GROUP BY sex;

```python
tips.groupby("sex").size()
```

Notice that in the pandas code we used size() and not count(). 
This is because count() applies the function to each column, 
returning the number of NOT NULL records within each.

```python
tips.groupby("sex").count()
```

SELECT day, AVG(tip), COUNT(*) FROM tips GROUP BY day;

```python
tips.groupby("day").agg({"tip": np.mean, "day": np.size})
```

SELECT smoker, day, COUNT(*), AVG(tip) FROM tips GROUP BY smoker, day;

```python
tips.groupby(["smoker", "day"]).agg({"tip": [np.size, np.mean]})
```
